```python
X_data = pd.read_csv('X_data.csv')
X_data
```

```python
y_data = pd.read_csv('y_data.csv')
y_data
```

```python
y_data = y_data['label']
y_data.head()
```

```python
X_data =X_data['title']
X_data.head()
```

```python
import re 
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk import WordNetLemmatizer
nltk.download('wordnet')
```

```python
Lemetizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(X_data)):
  sentence = re.sub('[^a-zA-Z]',' ',X_data[i])
  sentence = sentence.lower()
  sentence = nltk.word_tokenize(sentence)
  sentence = [word for word in sentence if word not in set(stopwords.words('english'))]
  sentence = [Lemetizer.lemmatize(word)for word in sentence]
  corpus.append(' '.join(sentence))
corpus
```

```python
len(corpus)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
dim = 10000
```

```python
onehot = []
for i in range(len(corpus)):
  onehot.append(one_hot(corpus[i],dim))
```

```python
print(onehot)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import LSTM,Dense,Embedding
from tensorflow.keras.models import Sequential
```

```python
padded = pad_sequences(onehot,padding='pre')
print(padded[0])
```

```python
padded.shape
```

```python
import numpy as np
X_final = np.array(padded)
y_final = np.array(y_data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.3, random_state = 42)
```

```python
X_train[1:5]
```

```python
model = Sequential()
model.add(Embedding(input_dim=dim,output_dim=20,input_length=padded.shape[1]))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred = model.predict_classes(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
